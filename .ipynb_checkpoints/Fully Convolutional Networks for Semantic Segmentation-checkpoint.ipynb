## 소개

- 모든 픽셀에 대한 prediction을 만드는 것이 중요단계
- 각 픽셀은 픽셀에 대응하는 객체나 영역의 클래스로 라벨링 됨


- FCN은 end to end, pixel-to-pixel로 semantic segmentation의 state of art 달성

- 픽셀 단위 예측을 위해 첫번째로, FCN을 엔드 투 엔드로, pretraining을 통해 FCN을 교육합니다.
- 우리의 모델은 분류망을 학습된 표현으로부터 fully conv과 fine-tunning 으로 재해석하여 최근의 분류 성공을 밀도 있는 예측으로 전달합니다. 이와는 대조적으로, 이전 작업에서는 pre-training없이 소규모 컨벤션(convnet)을 적용했습니다.

- 심층 feature 계층은 비선형 local-global 피라미드에서 위치와 semantic(픽셀의 디테일한 정보)을 인코딩합니다.

- 우리는 section 4.2에서 다룰 deep, coarse한 semantic 정보와 shallow, fine, appearance한 정보를 결합하는 feature 스펙트럼을 활용하기 위해 skip 구조를 정의합니다.



- patchwise training**하나의 이미지에서 객체가 존재하는 위치를 crop하여 하나의 patch로 만든 뒤 모델에 입력하는 training 방식**

# Fully convolutional networks



- convnet의 각 layer는 three-dimensional array(h x w x d)
- hxw = spatial dimensions
- d = channel dimension


- 가장 상위 계층의 location은 path-connected된 이미지의 location에 해당하며, 이를 receptive fields라고 합니다. -즉, conv한 결과 중 하나의 픽셀이 conv의 입력값에서 어느 크기의 영역에서 커버할 수 있는지를 결정. -> (deeplab 논문의 내용 인용)여러 convolution과 pooling 과정에서 디테일한 정보가 줄어들고 feature가 점점 추상화되는 것을 어느정도 방지


- CNN의 중요한 특성 중 하나는 입력 이미지에서 layer(convolution, pooling, activation function)를 통과시키더라도 위치 정보가 불변한다는 점입니다. 이를 논문에서는 translation invariant라고 말합니다.
- 단지 receptive fields에 해당하는 feature map들의 영역만 영향을 받습니다. 
- 몇 가지 기호에 대해서 언급하면  $x_{ij}$ 는 특정 layer의 (i,j) 위치에 해당하는 data vector라고 하겠습니다. 
- 이때 $x_{i,j}$가 vector인 이유는 feature map의 결과가 3차원의 형태로 나올 텐데 특정 위치에 해당하는 data의 이전 layer의 kernel의 개수와 동일하기 때문입니다. 그리고 $y_{i,j}$는 $x_{i,j}$ 에 의해 계산되는 결과라고 하겠습니다. 이를 수학적으로 표기하면 아래와 같습니다.


$$y_{ij} = f_{ks} (\{x_{si+δi,sj+δj}\}0≤δi,δj≤k)$$


- k는 커널 사이즈, s는 stride 또는 subsampling factor라고 하겠습니다.(stride의 크기에 따라 conv 출력값의 크기가 변하기 때문에 subsampling factor라고도 함)
- $f_{ks}$는 layer type을 결정합니다.: convolution or average pooling,  max pooling, activation function 등등

- 위 함수식에서 2개의 함수(kernel size, stride)를 합성하면 아래의 transformation rule을 만족합니다.


- 함수 합성을 적용하는 거는 전체 신경망의 앞에서 뒤로 나아가는 것을 의미합니다. 즉, data가 각종 layer를 통과하는 것을 의미합니다. 그런데 위의 f, g는 항상 nonlinear layer입니다. 따라서 data는 항상 nonlinear filter를 계산하게 됩니다. 이런 구조를 갖는 신경망을 fully convolutional network라고 정의합니다. 이 FCN은 아무 입력 이미지의 크기에 대해서 연산을 수행할 수 있고 resampling 과정을 통해서 입력 이미지에 대응되는 크기의 출력 이미지를 만들수 있습니다.

# Adapting classifiers for dense prediction

- FCN 이전에 유명한 신경망 구조를 보면. 대표적으로 LeNet, AlexNet, VGG 등등이 있다. 이 신경망들은 모두 반드시 고정된 입력 이미지의 크기를 입력받고, nonspatial한 출력을 얻습니다. 여기서 nonspatial 출력은 단순한 classification 결과를 의미합니다. 예를 들어 강아지, 고양이 등을 분류하기위해서 출력이 0, 1이 나오는 것입니다. 단순 scalar이기 때문에 nonspatial입니다. 위 예시의 fully connected layer(FCL)는 spatial coordinates를 버리게 됩니다.(Convolution layer로 진행시에는 위치정보를 유지하고 Flatten layer로 강제로 dense layer를 만들면ㅅ서 위치정보가 사라진다.) 
- 하지만 본 논문에서는 관점을 바꿉니다. FCL의 dense layer 역시 kernel을 통해 얻는 convolution layer로 바라볼 수 있다라는 것입니다. 그래서 아래의 그림과 dense layer였던 마지막 3개의 layer를 convolution layer로 바꿉니다. 이를 convolutionalization이라고 하였습니다

- classification을 fully conv으로 재해석하면 모든 크기의 입력에 대한 output map이 산출되지만, output dimension은 일반적으로 subsampling에 의해 감소됩니다.
- 이러한 coarse한 fully conv의 output은 입력의 크기로부터 출력 유닛의 receptive field의 픽셀 stride와 같은 factor로 인해 감소합니다.(즉, stride로 인한 feature map의 크기 감소를 말하는 것)


# 3.2. Shift-and-stitch is filter rarefaction

- https://stackoverflow.com/questions/40690951/shift-and-stitch-in-fcn를 참고해서 이 part를 설명하자면, 위의 고양이 강아지를 분류하는 네트워크 구조에서도 보이듯이 Upsampling을 사용하지 않은 경우 입력 이미지에 비해서 최종 출력 이미지의 해상도는 현저하게 낮습니다. 
- 예를 들어 100×100의 입력 이미지를 넣으면 10×10 크기의 최종 출력 이미지가 얻어진다고 가정을 하는 경우, 저희는 이를 semantic segmentation을 위해서 입력 이미지와 출력 이미지와의 크기를 동일하게 맞춰주어야합니다. 


- 그런데 만약 출력 이미지를 그대로 upsampling하게 되면 당연히 coarse한 feature map을 upsampling 하기 때문에 그만큼 정보의 손실도 많고 디테일하지 않습니다. 하지만 이러한 과정을 동일한 최종 출력 이미지를 가지고 약간 이동하고 다시 출력을 얻고 그 다음에 다시 조금 이동하고 출력을 얻고 반복합니다. 그러면 출력 이미지가 1개가 아니라 여러 개의 출력 이미지를 얻게 됩니다. 이렇게 얻은 출력 이미지의 집합을 더 고해상도의 이미지를 얻기 위해서 서로 결합하여 사용하게 됩니다.




- 2*2 max pooling layer로 stride=2로 feature map을 계산시 오른쪽의 결과를 얻게 됩니다.


- 밑에는 위의 input을 왼쪽으로 한칸 이동한 후 maxpooling을 진행합니다. 위와 다른 결과가 나오는 것을 알 수 있습니다.




- 같은 방식으로 위로, 오른쪽으로 한번씩 이동 후 max pooling을 진행하면 다른 결과를 얻습니다.

- 이제 위의 4가지 방법으로 얻은 feature를 합치면 다음과 같은 결과를 얻게 됩니다.


- 어디서 max pooling 연산이 적용되었는 지에 대한 위치 정보만 저장한다면 위와 같이 다시 입력 이미지와 같은 크기로 채울 수 있을 것입니다. 
- 하지만 이 알고리즘의 치명적은 단점은 계산 비용이 크다는 점입니다. 따라서 계산비용을 낮추기 위해서 본 논문에서는 위와 같은 upsampling을 위해서 트릭을 사용하였습니다. 



- s를 stride, $f_{ij}$를 convolution layer의 필터 가중치라고 하겠습니다. 다음 layer의 필터의 가중치를 만약 s가 i, j를 나눌 수 있다면$f'_{ij}=f_{i/s,j/s}$로 채우고 나누지 못한다면 0으로 채웁니다. 그러나 위의 shift-and-stitch와 완전히 동일한 결과는 얻을 수는 없을 것입니다. 

- 왜냐하면 오직 s에 따라서 채워지는 위치가 달라지기 때문이지요. 또한 shift-and-stitch와 다른 점은 max pooling을 사용하지 않고 학습가능한 가중치로 필터를 채운다는 점입니다. 따라서 이 upsampling하는 필터는 적절하게 학습까지 진행할 수 있다는 점이 차이점입니다. 하지만 본 논문에서는 이렇게 길게 설명하고 이 트릭을 사용하지 않았습니다. 왜냐하면 나중에 설명할 skip connection을 통해 구현한 것이 더 효과적이기 때문입니다.

# 3.3. Upsampling is backwards strided convolution

- 3.2절의 최종적인 목표는 낮은 해상도의 이미지를 높은 해상도의 이미지로의 upsampling이 목표였습니다. 이를 layer를 중심으로 다시 설명하면 coarse한 결과 이미지를 dense하게 픽셀을 채우는 것입니다. 이를 위한 대표적인 방법으로는 보간법(interpolation)입니다. 예를 들어 bilinear interpolation 같은 경우에는 $y_{ij}$에서 4개의 맞닿는 픽셀을 사용해서 보간을 할 수 있습니다.


- 본 논문에서는 backward convolution(deconvolution, transposed convolution)이라고 불리는 학습 기반 upsampling을 적용하였습니다. 이렇게 backward convolution과 activation function을 연결하여 nonlinear upsampling을 학습하게 됩니다.

# 3.4. Patchwise training is loss sampling


- 본 논문에서는 patchwise training이라는 개념을 언급하는데,  patchwise training이란 모델을 training 할 때 전체 이미지를 넣는 것이 아니라 객체의 주변을 분리한 서브 이미지 즉 patch image를 사용하는 방법을 의미합니다. 여기서 이미지의 부분을 학습하는 patch training과 전체 이미지를 학습하는 fully convolution training은 서로 동일하다고 설명하고 있습니다. 
- patchwise training에서 sampling은 class 불균형을 교정하고 dense patch의 공간 상관관계를 완화시킬 수 있다
- fully conv training에서 loss에 가중치 부여함으로써 class 균형을 달성할 수도 있다. 
- 또한 본 논문에서는 fully convolution training이 더 계산적으로 유리하다고 언급하였습니다.(section 4.2 참고)

# 4. Segmentation Architecture

- 본 논문에서는 ILSVRC의 classifier를 FCN에 추가시키고, pixel 별 loss 계산와 upsampling을 dense prediction에 보충합니다. fine-tuning을 통해서 segmentation을 위한 학습을 진행하고, skip connection을 coarse하고 semantic한 layer와 local하고 appearance한 layer 사이에 추가하였습니다. skip connection 추가한 네트워크 구조를 skip architecture라고 하는데 이 skip architecture는 출력의 semantic과 공간 정밀도를 다듬기위해 end-to-end로 학습하게 됩니다.


- 그리고 본 논문에서 사용한 데이터 셋은 PASCAL VOC 2011 segmentation challenge에서 사용된 데이터입니다. 학습을 할 때는 pixel 별 multinomial logistic loss를 사용하고 검증을 위한 metric으로 mean pixel intersection over union(meanIoU)를 사용해 평균이 배경을 포함한 모든 클래스를 나타냅니다.

- 무시하는 픽셀의 학습은 ground truth에서 모호하거나 어려운 픽셀을 배제합니다

-  DAG 네트워크는 coarse하고 high layer 정보와 미세하고 low layer 정보를 결합하는 방법을 배웁니다. -> skip connection

- pooling 및 prediction layer는 상대적인 spatial coarseness을 드러내는 grid로 표시되고 중간 layer는 수직선으로 표시됩니다. 


- 첫 번째 행(FCN-32s): 섹션 4.1에 설명된 NAT의 단일 스트림 네트워크는 stride=32로 예측을 한 번에 픽셀로 되돌립니다. 
- (FCN-16s): output layer과 pooling 4의 layer의 예측(feature map)을 모두 결합하고 stride=16로  높은 수준의 semantic 정보를 유지하면서 보다 세부적인 정보를 예측할 수 있습니다. 
- 세 번째 행(FCN-8s): 3번 풀과 prediction의 합은 stride=8로 fcn-16s보다 더 높은 성능을 나타냅니다.




- 이 그림은 IoU를 간단하게 표현한 것입니다. 실제 mask이미지와 예측된 mask 이미지 사이에 픽셀 단위로 얼마나 겹치는 지를 비율로 표현해주는 metric으로 쓰입니다. 1에 가까울 수록 좋은 성능을 지녔음을 의미합니다.

# 4.1. From classifier to dense FCN

- 제 본 논문에서 말하는 FCN을 만들기 위해서 사용한 네트워크는 AlexNet(ILSVRC2012), VGG(ILSVRC2014), GoogLeNet(ILSVRC2014)를 사용하였습니다. 특히 VGG같은 경우에는 VGG16이 있고 VGG19가 있는데 본 논문에서는 VGG16을 사용하였습니다. 이 네트워크들을 1절에서 언급한 convolutionalozation 과정을 통해서 전부 FCN으로 만들게 됩니다. 이 과정에서 이전에 언급했다 싶이 1×1 합성곱 layer를 사용하였습니다. 그리고 PASCAL 데이터셋의 클래스가 21개(배경을 포함)이기 때문에 원래 Fully connected layer였던 부분은 21개의 채널을 가지는 layer로 바뀌게 됩니다.





- 다음의 그림은 위에서 언급한 네트워크들을 FCN으로 바꾸고 성능을 측정한 결과입니다. mean IU는 IoU와 동일한 metric인데 VGG16일 때 가장 성능이 좋은 것을 볼 수 있습니다. 하지만 VGG16의 forward time, parameter가 많아서 네트워크 자체가 무거울 것 같습니다. 그래도 전체적으로 봤을 때 성능이 엄청 좋은 것 같지는 않습니다. 그래서 본 논문에서 skip architecture를 제시합니다. 

# 4.2. Combining what and where

- 위의 네트워크 구조를 다시 보면 먼저, FCN-16은 pool4 layer에서 1×1 합성곱 layer를 적용하고, 마지막 layer에서 x2 upsampling을 한 뒤에 합칩니다. 마지막 layer에서 x32배 upsampling을 하지 않고 x16배 upsampling하면 입력 이미지의 크기와 동일한 출력 이미지를 얻을 수 있습니다. 
- 그리고 FCN-8은 FCN-16에서 x16배 upsampling하기 전 결과에서 x2 upsampling을 한 뒤에 pool3 layer에 1×1 합성곱 layer를 적용하여 하나로 합칩니다. 그리고 x8배 upsampling을 하게 되는 것입니다. 
- 위의 과정에서 단순히 마지막 layer에서 x32배 upsampling을 한 것은 FCN-32라고 하겠습니다. 여기서 1x1 conv의 목적은 coarse한 layer이 fine layer와 elementwise sum를 하기 위해 channel을 맞춰주는 역할이라 생각할 수 있습니다. 그리고 coarse한 layer와 fine layer를 합치는 것은 논문에서는 모델이 global 구조를 포함하면서 local한 prediction을 만든다고 언급합니다. 


- 정리하면, coarse한 layer가 가지는 위치정보와 fine layer가 가지는 디테일한 정보를 합쳐주기 때문에 segmentation에서 고질적인 문제인 localization과 semantic information 문제를 해결합니다.



- 각각의 skip architecture의 결과를 시각화하여 확인해보겠습니다.


- 위의 표는 skip connection을 어디에서 했는 지에 따라서 성능이 어떻게 차이가 나는 지 보여주고 있습니다. 실제로 위에서 언급했듯이 fine layer에는 입력 이미지의 세세한 정보가 많이 담겨있을 것입니다. 그래서 convolution layer를 진행하면서 조금씩 손실되는 이미지의 정보를 skip connection을 통해서 손실을 막아주는 역할을 하는 것이죠. 따라서 FCN-16보다 FCN-8이 더 좋은 성능을 보이는 이유는 16보다 더 higher 에 속한 fine layer에서 skip connection을 주었기 때문에 더 디테일한 이미지의 정보를 얻을 수 있었기 때문이라고 해석할 수 있을 것 같습니다.

- pooling layer의 stride를 줄이는 것이 보다 정확한 예측을 얻을 수 있는 가장 간단한 방법입니다. 그러나 그렇게 하는 것은 VGG16 기반 네트워크에 문제가 있습니다.
- pool5에서 stride=1로, fully conv를 14*14로 커널 크기로 컨볼루션을 진행하면 receptive field의 크기를 유지하게 됩니다.하지만 이러한 방법은 너무 큰 커널 크기로 학습이 어려워지는걸 발견합니다.
- 추가로 pool5위의 conv층을 더 작은 필터로 변환해봅니다. 하지만 역시나 성능이 좋지 않다는 걸 확인합니다.
- 마지막으로 위에서 소개한 shift-and-stitch를 사용해봤지만 더 나쁜 성능을 보입니다.


# 4.3. Experimental framework

- Optimization: 논문에서는 momentum으로 SGD를 사용해 학습합니다. 20개 이미지의 미니배치 크기 및 고정된 학습률을 사용합니다. momentum=0.9로, weight decay는 $5^{−4} or 2^{−4}$ 로 하고 bias 학습을 두배로 늘립니다. 그리고 논문에서는 훈련이 학습률에만 민감하다는 것을 알게 되었지만. 랜덤 초기화가 더 나은 성능과 더 빠른 융합을 제공하지 못했기 때문에 class scoring링 계층을 제로 초기화합니다. original classifier net이 사용된 위치에 dropout이 포함되었습니다.
