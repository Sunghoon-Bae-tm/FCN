```python
def newshape(x):
    if len(x.shape) ==4:
        N,H,W,C=x.shape
        
        
        # 제 1영역
        x1= x[:,0:H:2,0:W:2] #0:마지막층:2
    
        #제 2영역
        x2= x[:,0:H:2,1:W:2] #1:마지막층:2

        #제 3영역
        x3= x[:,1:H:2,0:W:2]

        #제 4영역
        x4= x[:,1:H:2,1:W:2]
        
    elif len(x.shape) == 3:
        H, W, C= x.shape                   
        
        
        # 제 1영역
        x1= x[0:H:2,0:W:2] #0:마지막층:2
    
        #제 2영역
        x2= x[0:H:2,1:W:2] #1:마지막층:2

        #제 3영역
        x3= x[1:H:2,0:W:2]

        #제 4영역
        x4= x[1:H:2,1:W:2]
        
    else:
        H,W = x.shape
      
        
        # 제 1영역
        x1= x[0:H:2,0:W:2] #0:마지막층:2
    
        #제 2영역
        x2= x[0:H:2,1:W:2] #1:마지막층:2

        #제 3영역
        x3= x[1:H:2,0:W:2]

        #제 4영역
        x4= x[1:H:2,1:W:2]
    
    
    return x1, x2, x3, x4  
    
```

```python
def vh_concat3(x1, x2):
    even = tf.keras.layers.concatenate([x1[:, 0::2], x2[:, 0::2]], axis=1)
    odd = tf.keras.layers.concatenate([x1[:, 1::2], x2[:, 1::2]], axis=1)
    
    return tf.keras.layers.concatenate([even, odd], axis=1)
    
def upshape3(v, h):
    even = tf.keras.layers.concatenate([v[:, 0::2, :], h[:, 0::2, :]], axis=0)
    odd = tf.keras.layers.concatenate([v[:, 1::2, :], h[:, 1::2, :]], axis=0)
    
    print(even)
    print(odd)
    return tf.keras.layers.concatenate([even, odd], axis=0) 
    
```

```python
def vh_concat4(x1, x2):
    even = tf.keras.layers.concatenate([x1[:, :, 0::2], x2[:, :, 0::2]], axis=2)
    odd = tf.keras.layers.concatenate([x1[:, :, 1::2], x2[:, :, 1::2]], axis=2)
    
    return tf.keras.layers.concatenate([even, odd], axis=2)
    
def upshape4(v, h):
    #even = tf.keras.layers.concatenate([v[:, 0::2, :], h[:, 0::2, :]], axis=1)
    #odd = tf.keras.layers.concatenate([v[:, 1::2, :], h[:, 1::2, :]], axis=1)
    even = tf.keras.layers.concatenate([v[:, 0::2, :], h[:, 0::2, :]], axis=1)
    odd = tf.keras.layers.concatenate([v[: ,1::2, :], h[: ,1::2, :]], axis=1)
    
    print(even)
    print(odd)
    return tf.keras.layers.concatenate([even, odd], axis=1) 
    
```

```python
def upshape(x1,x2,x3,x4):
    x1_even = x1[:, :, 0::2][:, :, 0::2]
    x1_subeven = x1[:, :, 0::2][:, :, 1::2]
    x1_odd = x1[:, :, 1::2][:, :, 0::2]
    x1_subodd = x1[:, :, 1::2][:, :, 1::2]

    x2_even = x2[:, :, 0::2][:, :, 0::2]
    x2_subeven = x2[:, :, 0::2][:, :, 1::2]
    x2_odd = x2[:, :, 1::2][:, :, 0::2]
    x2_subodd = x2[:, :, 1::2][:, :, 1::2]


    x3_even = x3[:, :, 0::2][:, :, 0::2]
    x3_subeven = x3[:, :, 0::2][:, :, 1::2]
    x3_odd = x3[:, :, 1::2][:, :, 0::2]
    x3_subodd = x3[:, :, 1::2][:, :, 1::2]

    x4_even = x4[:, :, 0::2][:, :, 0::2]
    x4_subeven = x4[:, :, 0::2][:, :, 1::2]
    x4_odd = x4[:, :, 1::2][:, :, 0::2]
    x4_subodd = x4[:, :, 1::2][:, :, 1::2]

    even = tf.keras.layers.concatenate([x1_even, x2_even, x1_odd, x2_odd, x1_subeven, x2_subeven, x1_subodd, x2_subodd], axis=2)
    odd = tf.keras.layers.concatenate([x3_even, x4_even, x3_odd, x4_odd, x3_subeven, x4_subeven, x3_subodd, x4_subodd], axis=2)

    j=[]
    p=[]
    f=[]
    B, H, W, C= even.shape
    for v in range(H):
        for i in range(B):
            new1=tf.expand_dims(even[:,v, :, :][i], axis=0)
            new2= tf.expand_dims(odd[:,v, :, :][i], axis=0)
            j.append(tf.keras.layers.concatenate([new1, new2] ,axis=0))
        print(len(j))
    
    for h in range(B):
        p.append(tf.keras.layers.concatenate([j[h], j[h+B], j[h+B*2], j[h+B*3]] ,axis=0))

    for c in range(len(p)):
        f.append(tf.expand_dims(p[c], axis=0))
    
    if len(p) == 4:
        return(tf.keras.layers.concatenate([f[0], f[1], f[2], f[3]],axis=0))
    
    elif len(p) == 8:
        return(tf.keras.layers.concatenate([f[0], f[1], f[2], f[3], f[4], f[5], f[6], f[7]],axis=0))
    
    else:
        print("error")
    
#for h in range()
#tf.keras.layers.concatenate([p[0], p[1], p[2], p[3]] ,axis=0)
```
